This project work deals with a method of matrix decomposition, the so-called singular value decomposition (SVD) and its application to image compression.

SVD is an orthogonalization method for non-square matrices. This method can be used to determine the rank of a matrix and to solve linear balancing problems.

The SVD of a complex m√ón matrix A with rank r is a product of the form $$A = U \Sigma V^T$$

```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import scipy.linalg as ln
import numpy as np
import os
from PIL import Image
from math import log10, sqrt 
plt.rcParams['figure.figsize'] = [16, 8]
# Import image
A = imread(os.path.join("../input/image-htw/htw-berlin_emas-zertifiziert.jpg"))
X = A.dot([0.299, 0.5870, 0.114]) # Convert RGB to grayscale
img = plt.imshow(X)
# Konvert to gray
img.set_cmap('gray')
plt.axis('off')
plt.show()
# Call image size
print('Bildgr√∂√üe: %s'%str(X.shape))
```

```python
# split the matrix into U, S, VT
U, S, VT = np.linalg.svd(X,full_matrices=False)
S = np.diag(S)
m = 800 # Image's width
n = 1200 # Image's height
j = 0
# Try compression with different k:
for k in (5,10, 20, 100,200,400,500):
    # Original size of the image
    originalSize = m * n 
    # Size after compressed
    compressedSize = k * (1 + m + n) 
    # A√º√ºrpximation of A
    Xapprox = U[:,:k] @ S[0:k,:k] @ VT[:k,:]
    plt.figure(j+1)
    j += 1
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    
    plt.axis('off')
    plt.title('k = ' + str(k))
    plt.show() 
    print('Originalgr√∂√üe:')
    print(originalSize)
    print('komprimierte Gr√∂√üe / Originalgr√∂√üe:')
    # Estimate the compression rate
    ratio = compressedSize * 1.0 / originalSize
    print(ratio)
    print('Compression rate is ' + str( round(ratio * 100 ,2)) + '%' )  
    # Estimate MQA
    x= X.astype("float")
    y=Xapprox.astype("float")
    err = np.sum((x - y) ** 2)
    err /= float(X.shape[0] * Xapprox.shape[1])
    print('Die Mittlere quadratische Abweichung betr√§gt '+ str(round( err)))
    max_pixel = 255.0
    # Estimate Signal Noise Ratio
    srv = 20 * (log10(max_pixel / sqrt(err)))
    print('Signal-Rausch-Verh√§ltnis betr√§gt '+ str(round(srv)) +'dB')
```

From the results we can see that image compression can be achieved by choosing an appropriate value of k. 
* If k = 5 or the image is reconstructed only from the first 5 eigenvalues of the matrix , the image is so blurred that we can hardly recognize its content. 
* Moreover, the compression rate is only 1% with k = 5, which means that the compressed image is only 1% of the original image and a lot of memory is saved. But it results in a large measure of degradation compared to the original image when the MQA is so large, which clarifies why the image is so blurry.
