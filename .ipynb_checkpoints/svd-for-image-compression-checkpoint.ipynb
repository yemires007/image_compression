# SVD for Image Compression

```python
# Common imports!
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img


np.random.seed(42)
```

## Importing the Image

```python
# We need an image to work with!
the_image = img.imread('../input/call-of-duty-advanced-warfare-image/codadwrf.png')
the_image.shape
```

```python
plt.figure(figsize = (16, 8))
plt.imshow(the_image)
plt.axis('off')
plt.show()
```

```python
# To make it easy, I multiply & reshape the 2nd and 
# 3rd dims of the image to to make it a 2d numpy nd array

the_image = the_image.reshape(1600, 2560 * 4) 
the_image.shape
```

# Singular Value Decomposition 
In Linear Algebra, any given **m** $\times$ **n** matrix **M** can be factorized into the dot product of three other matrices:
    
   1. an $m \times m$ orthogonal matrix; an orthogonal matrix 
   2. an $m \times n$ diagonal matrix; a diagonal matrix
   3. $n \times n$ orthogonal matrix; a rotation matrix

$M = U \cdot \Sigma \cdot V^{T}$



## Factoring the Matrix of the Image

```python
# I use SVD to decompose the image, select the most importat components 
# and put them back to compress it.

U, s, Vt = np.linalg.svd(the_image, full_matrices = False)
S = np.diag(s)
```

```python
# The following code compresses the image with given number of r/components in the dims_ array

dims_ = (5, 10, 20, 30, 50, 100, 200, 250)
comp_imgs = np.array([U[:, :r] @ S[0: r, :r] @ Vt[: r, :] for r in dims_])
  
n_rows = 2
n_cols = 4
plt.figure(figsize=(n_cols * 5, n_rows * 5))
for row in range(n_rows):
    for col in range(n_cols):
        index = n_cols * row + col
        plt.subplot(n_rows, n_cols, index + 1)
        plt.imshow(comp_imgs[index].reshape(1600, 2560, 4), interpolation="nearest")
        plt.axis('off')
        plt.title("r = " + str(dims_[index]), fontsize=12)
plt.subplots_adjust(hspace = -0.5)
plt.show()
```

**Note:** as you can see, the images with the smaller selected number of r/components get more compressed and lose a lot of useful information, and as I increase the r/components the image tends to become more clear, i.e., since we tend to add more information from the original image.

```python
# To not compress the image based on a given random number r/components, 
# we are required to do some math & statistics as following:

#'''
#    1. The right plot  "Component Importance" shows the  amount 
#       of variance each  component carries. The first component 
#       preserves the highest variance, then the second, and  so 
#       on.
#    2. The left plot shows the explained variance as a function
#       of numbers, using this technique you can choose how much
#       the image, relative to its original size gets compressed.
#'''

plt.figure(figsize = (15, 5))
plt.subplot(121)
plt.semilogy(np.diag(S), 'r-', linewidth = 1.9)
plt.xlabel('Number of PC')
plt.ylabel('Log Function', rotation = 90)
plt.title('Component Importance')
plt.grid(True)

plt.subplot(122)
plt.plot((np.cumsum(np.diag(S)))/(np.sum(s)), 'r+-', linewidth = 1)
plt.title('Explained Variance as Function of Numbers')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio', rotation = 90)
plt.grid(True)

plt.show()
```

```python
# Here, I decided to compress the image with 90% of its
# explained variance ratio:

explained_variance_ratio = np.cumsum(np.diag(S)/np.sum(s))
precentage = 0.90
d = np.argmax(explained_variance_ratio >= 0.90) + 1
explained_variance_ratio[d - 1]
```

```python
image_with_90_variance = U[:, :d] @ S[: d, : d] @ Vt[:d, :]

plt.figure(figsize = (16, 8))
plt.imshow(image_with_90_variance.reshape(1600, 2560, 4))
plt.axis('off')
plt.title('Compressed with 90% of Variance')
plt.show()
```
